```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
import glob
import statsmodels.api as sm
```

```python
dataPaths = glob.glob("./*.csv")
dataDict = {}
delta = 0.5
for data in dataPaths:
    dataDict[data[2:].split(".")[0]] = pd.read_csv(data, header=None)
```

```python
plt.plot(dataDict["S1_f"])
plt.plot(dataDict["gate1"])
plt.show()
```

```python
plt.plot(dataDict["S1"])
plt.plot(dataDict["gate1"])
plt.show()
```

```python
fig, (ax1, ax2) = plt.subplots(2, 1, figsize=(12, 6))
sm.graphics.tsa.plot_acf(dataDict["S1_f"], lags=40, ax=ax1)
sm.graphics.tsa.plot_pacf(dataDict["S1_f"], lags=40, ax=ax2)
plt.show()
```

```python
model = sm.tsa.ARIMA(dataDict["S1_f"], exog=dataDict["gate1"], order=(10, 1, 5))
results = model.fit()
```

```python
p_values = range(0, 15)
q_values = range(0, 15)
# i_values = range(0, 1)

best_aic = float("inf")
best_order = (0, 0, 0)

for p in p_values:
    for q in q_values:
        # for i in i_values:
        try:
            model = sm.tsa.ARIMA(dataDict["S1_f"], exog=dataDict["gate1"], order=(p, 1, q))
            results = model.fit()
            aic = results.aic
            if aic < best_aic:
                best_aic = aic
                best_order = (p, i, q)
        except:
            continue

print(f'Best AIC: {best_aic}')
print(f'Best Order: {best_order}')
```

```python
model = sm.tsa.ARIMA(dataDict["S1_f"], exog=dataDict["gate1"], order=(5, 1, 5))
results = model.fit()
print(results.summary())
```

```python
# Tahminlerde bulunun
i = 100
j = i+200

if len(dataDict["gate1"].iloc[i:j]) == j-i:
    print("results are not true")

forecast = results.predict(start=i,exog=dataDict["gate1"].iloc[i:j], end=j, dynamic=False, typ='levels')

plt.figure(figsize=(12, 6))
plt.plot(dataDict["t"], dataDict["S1_f"], label='output', color='blue')
plt.plot(dataDict["t"], dataDict["gate1"], label='input', color='orange')
plt.plot(np.arange(i, j+1, 1), forecast, label='predict', color='red')
plt.legend()
plt.title('ARIMA Forecasts')
plt.xlabel('Tarih')
plt.ylabel('Y Deikeni')
plt.show()
```

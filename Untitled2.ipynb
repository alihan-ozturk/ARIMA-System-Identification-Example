```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from scipy.signal import lfilter
from statsmodels.tsa.arima.model import ARIMA
```

```python
# Generate time-delayed data with noise
np.random.seed(0)
t = np.arange(0, 100)
delayed_data = 0.5 * np.sin(0.1 * t - 0.5)  # Example time-delayed system
noise = np.random.normal(0, 0.2, len(t))
noisy_data = delayed_data + noise
```

```python
# Define the low-pass filter coefficients (you may need to adjust these)
b = [0.01] * 10  # Example filter coefficients
a = 1

# Apply the low-pass filter to reduce noise
filtered_data = lfilter(b, a, noisy_data)
```

```python
plt.figure(figsize=(10, 6))
plt.plot(t, delayed_data, label='Original Data', color='green')
plt.plot(t, noisy_data, label='Noisy Data', color='red', alpha=0.5)
plt.plot(t, filtered_data, label='Filtered Data', color='blue')
plt.legend()
plt.title('Original, Noisy, and Filtered Data')
plt.xlabel('Time')
plt.ylabel('Value')
plt.show()
```

```python
# Create a DataFrame with the filtered data
df = pd.DataFrame({'Time': t, 'Filtered_Data': filtered_data})

# Fit an ARIMA model to the filtered data
model = ARIMA(df['Filtered_Data'], order=(4, 0, 1))  # Replace p, d, and q with appropriate values
results = model.fit()
```
